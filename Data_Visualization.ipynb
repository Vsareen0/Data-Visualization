# Data Visulization

```python
import pandas as pd
```

The next step is to read the csv file as dataframe. **DataFrame** is a 2-dimensional labeled data structure. The Dataframe object will be called as df

```python
df = pd.read_csv("parks.csv")
```

To Visualize the dataset we will use `head()` of Dataframe. The `head()` returns the first 5 rows of the dataset. You can control the rows to be displayed by passing a valid number.

```python
df.head(10)
```

Now we can index the rows using `iloc[]` method. Just pass in the row number to iloc method. **iloc** stands for INDEX LOCATION

```python
df.iloc[2]
```

The `loc()` method is used to locate and visualize multiple rows

```python
df.iloc[[2,1,0]]
```

```python
df[5:13]
```

Indexing Columns 

```python
df['State'].head(3)
```

```python
df.State
```

In the above cell, you have displayed the column State, but if any column has a space its name the you need to take care of it else an error would be generated. To do cleaning, we can use: 

```python
df.columns = [col.replace(' ','_').lower() for col in df.columns]
print(df.columns)
```

To perform queries or apply condition to visualize data use `df[df.column_name with some condition]`

```python
df[['state','acres']][:3]
```

```python
df[df.state == 'WA'].head(2)
```

```python
df[(df.latitude > 60)|(df.acres > 10**6)].head(3)
```

```python
df[(df.acres < 40000) | (df.latitude < 30)].head()
```

```python
df[df['park_name'].str.split().apply(lambda x:len(x) == 3)].head()
```

```python
df[df.state.isin(['WA','CA','OR'])]
```
